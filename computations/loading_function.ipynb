```python
import random

critical_wait_time = 240

def random_current_wait_times(start, end):
    return random.randint(start, end)

start_range = 15
end_range = 240

def random_current_capacity(minmum, maximum):
    return random.randint(minmum, maximum)

min_capacity = 15
max_capacity = 100

```

```python
import pandas as pd

file_path = "site_dataset.csv"

df = pd.read_csv(file_path)
```

```python
def calculate_load_score(current_capacity, department_total_capacity, current_wait_time, critical_wait_time):
    load_score = 0
    load_score = ((current_capacity / department_total_capacity) * 100) + ((current_wait_time / critical_wait_time) * 100)
    return load_score

for index, row in df.iterrows():
    current_wait_time = random_current_wait_times(start_range, end_range)
    current_capacity = random_current_capacity(min_capacity, max_capacity)

    departments = row['Department'].split("|")
    department_capacities = row['Department Capacity']

    if isinstance(department_capacities, str):
        department_capacities = list(map(int, department_capacities.split("|")))
    else:
        department_capacities = [department_capacities]

    print("Site Code: ", row['Site Code'])
    print("Site Address: ", row['Site Address'])

    for i, department in enumerate(departments):
        load_score = calculate_load_score(current_capacity, department_capacities[i], current_wait_time, critical_wait_time)
        print(f"Department Name: {department}")
        print(f"Load Score:  {load_score:.2f}%")

        if load_score <= 70:
            print("Status: Departement is operating effeciently")
        elif 71 <= load_score <= 85:
            print("Status: Department is approaching capacity; initiate patient redirection to other departments or A&E sites.")
        else:
            print("Status: Department is Overloaded")

    print("\n")

```
